### Correcting the Spelling mistakes in the State columns

```python
df3['State'] = df3['State'].str.replace('*', '')
df3['State'] = df3['State'].str.replace('Himanchal', 'Himachal')
df3['State'] = df3['State'].str.replace('Telengana', 'Telangana')
df3['State'] = df3['State'].str.replace('Karanataka', 'Karnataka')
df3['State'] = df3['State'].str.replace('Dadra and Nagar Haveli', 'Dadra and Nagar Haveli and Daman and Diu')
df3['State'] = df3['State'].str.replace('Daman & Diu', 'Dadra and Nagar Haveli and Daman and Diu')
# df3['State/UnionTerritory'] = df3['State/UnionTerritory'].str.replace('', '')
```

```python
df3['State'] = df3['State'].str.replace('Dadra and Nagar Haveli and Daman and Diu and Daman and Diu', 'Dadra and Nagar Haveli and Daman and Diu')
```

### Grouping the datasets to view Statewise information

```python
df1_grouped = df1.groupby(['State']).sum().reset_index()
```

```python
df1_grouped.head()
```

```python
df2_selected = df2.loc[:, ['Date', 'State', 'TotalDoses',
                           'Covaxin', 'CoviShield',
                           'Sputnik', 'TotalVaccinated']]
```

```python
# Grouping the Statewise information of the vaccines used
df2_grouped = df2_selected.groupby(['State']).sum().reset_index()
df2_grouped.dropna()
df2_grouped = df2_grouped[df2_grouped['State'] != 'India']
#df2_grouped_1 = df2_grouped[df2_grouped['State'] == 'India'].reset_index(drop = True)
```

```python
df2_grouped.head()
```

```python
# Grouping the Statewise Death and Cured information
df3_grouped = df3.groupby(['State'])['Cured','Deaths','Confirmed'].sum().reset_index()
```

```python
df3_grouped.head()
```

### Merging the Datasets

```python
covid_grouped = df1_grouped.merge(df2_grouped, how='outer', on='State')
```

```python
covid_grouped = covid_grouped.merge(df3_grouped, how='outer', on='State')
```

```python
covid_grouped.head()
```

### Plotting the Data

```python
df1.columns
```

```python
plt.figure(figsize=(15,15))
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain')
plt.grid(True)

plt.title('Total number of Samples Collected (Statewise)')
plt.xlabel('Dates')
plt.ylabel('Total Samples Collected')

sns.lineplot(data=df1,
             x='Date',
             y='TotalSamples',
             hue='State',
             ci=0,
             palette='YlGnBu')

plt.show();
```

```python
plt.figure(figsize=(18,7))
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain')
plt.grid(True)

plt.title('Total number of Confirmed Covid-19 cases (Statewise)')
plt.xlabel('State Name')
plt.ylabel('Confirmed cases')

sns.barplot(data=df3,
             x='State',
             y='Confirmed',
             ci=0,
             palette='YlGnBu')

plt.show();
```

```python
plt.figure(figsize=(20,7))
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain')
plt.grid(True)

plt.title('Total number of vaccinations done (Statewise)')
plt.xlabel('State Name')
plt.ylabel('Number of Vaccines administered')

sns.barplot(data=df2_grouped,
            x='State',
            y='TotalDoses',
            ci=False,
            palette='ocean_r')

plt.show();
```

```python
bar_state = df2_grouped.iloc[:,0]
bar_covishield = df2_grouped.iloc[:,3]
bar_covaxin = df2_grouped.iloc[:,2]
bar_sputnik = df2_grouped.iloc[:,4]
indx = np.arange(len(df2_grouped))
```

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
plt.ticklabel_format(style='plain')

plt.xticks(indx, bar_state)
plt.title('Vaccination by Type (Statewise)')
plt.xlabel('State Name')
plt.ylabel('Number of Vaccines administered')

# plt.legend

graph_covishield = plt.bar(x=indx, height=bar_covishield)
graph_covaxin = plt.bar(x=indx, height=bar_covaxin, bottom=bar_covishield)
graph_sputnik = plt.bar(x=indx, height=bar_sputnik, bottom=bar_covaxin+bar_covishield)
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=0)
plt.ticklabel_format(style='plain')
plt.grid(True)

plt.title('Cured vs Deaths')
plt.xlabel('Date')
plt.ylabel('Count')

sns.lineplot(data=df3, x='Date', y='Deaths', ci=False)
sns.lineplot(data=df3, x='Date', y='Cured', ci=False)

plt.show()
```
