```python
Titanic['Pclass'].replace({'1': '1st Class', '2': '2nd Class', '3': '3rd Class'}, inplace=True)
```

```python
plt.figure(figsize=(12,12))

plt.subplot(2, 2, 1)
plt.title('Count based on Gender')
plt.xlabel('Gender')
sns.countplot(x='Gender',
              data=Titanic,
              palette='flare',
              order=['female', 'male']).set_xticklabels(['Female',
                                                         'Male'])
plt.subplot(2, 2, 2)
plt.title('Embarked')
plt.xlabel('Port Name')
sns.countplot(x='Embarked',
              data=Titanic,
              palette='flare').set_xticklabels(['Southampton',
                                                  'Cherbourg',
                                                  'Queenstown'])
plt.subplot(2, 2, 3)
plt.title('Count based on Class')
plt.xlabel('Passenger Class')
sns.countplot(x='Pclass',
              data=Titanic,
              palette='flare').set_xticklabels(['1st Class',
                                                '2nd Class',
                                                '3rd Class'])
plt.subplot(2, 2, 4)
plt.title('Count of Passenger Survived')
plt.xlabel('Surived vs Died')
sns.countplot(x='Survived',
              data=Titanic,
              palette='flare', # YlGnBu
              order=[1, 0]).set_xticklabels(['Survived',
                                             'Died'])

plt.show();
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2, 2, 1)
plt.grid(False)
plt.title('Age-wise distribution')
plt.xlabel('Age')
plt.ylabel('Count')
sns.histplot(data=Titanic,
             x='Age',
             hue='Gender',
             kde=False)
plt.subplot(2, 2, 2)
plt.grid(True)
plt.title('Age-wise distribution')
plt.xlabel('Age')
plt.ylabel('Count')
sns.kdeplot(data=Titanic,
             x='Age',
             hue='Gender')
plt.subplot(2, 2, 3)
plt.grid(False)
plt.title('Age-wise distribution')
plt.xlabel('Age')
plt.ylabel('Count')
sns.histplot(data=Titanic,
             x='Age',
             hue='Pclass',
             kde=False)
plt.subplot(2, 2, 4)
plt.grid(True)
plt.title('Age-wise distribution')
plt.xlabel('Age')
plt.ylabel('Count')
sns.kdeplot(data=Titanic,
             x='Age',
             hue='Pclass')

plt.show();
```

```python
plt.figure(figsize=(15,5))
plt.xticks(np.arange(0,501,25))
plt.ticklabel_format(style='plain')
plt.grid(False)

plt.title('Fare-wise distribution')
plt.xlabel('Fare')
plt.ylabel('Count')

sns.histplot(data=Titanic,
             x='Fare',
             kde=True,
             bins=100)

plt.show();
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1)
plt.grid(True)
plt.title('Fare vs Class vs Gender (Total Amount)')
plt.xlabel('Class')
plt.ylabel('Fare')
sns.barplot(data=Titanic,
             x='Pclass',
             y='Fare',
             hue='Gender',
             ci=0, estimator=np.sum,
             palette='ocean').set_xticklabels(['1st Class',
                                               '2nd Class',
                                               '3rd Class'])
plt.subplot(1, 2, 2)
plt.grid(True)
plt.title('Fare vs Class vs Gender (Average Amount)')
plt.xlabel('Class')
plt.ylabel('Fare')
sns.barplot(data=Titanic,
             x='Pclass',
             y='Fare',
             hue='Gender',
             ci=0, estimator=np.mean,
             palette='ocean').set_xticklabels(['1st Class',
                                               '2nd Class',
                                               '3rd Class'])

plt.show();
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1)
plt.grid(False)
plt.title('Class vs Gender (Count)')
plt.xlabel('Class')
plt.ylabel('Count')
sns.countplot(data=Titanic,
             x='Pclass',
             hue='Gender',
             palette='ocean').set_xticklabels(['1st Class',
                                                '2nd Class',
                                                '3rd Class'])
plt.subplot(1, 2, 2)
plt.grid(False)
plt.title('Age vs Class vs Gender (Average Age)')
plt.xlabel('Class')
plt.ylabel('Age')
sns.barplot(data=Titanic,
             x='Pclass',
             y='Age',
             hue='Gender',
             ci=0, estimator=np.mean,
             palette='ocean').set_xticklabels(['1st Class',
                                                '2nd Class',
                                                '3rd Class'])

plt.show();
```
